```python
test_data = pd.read_csv('GoldenOak_Test.csv')
test_data
```

```python
x_test = test_data.drop('Defaulted', axis = 1)
y_test = test_data['Defaulted']
```

```python
def confusion_matrix(data, actual_values, model):
    predictions = model.predict(data)
    bins = np.array([0, 0.5, 1])
    cm = np.histogram2d(actual_values, predictions, bins = bins)[0]
    accuracy = (cm[0, 0] + cm[1, 1])/cm.sum()
    return cm, accuracy
```

```python
confusion_matrix(x_test, y_test, model)
```
